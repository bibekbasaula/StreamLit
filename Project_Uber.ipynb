```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('ncr_ride_bookings.csv')
df.head()
```

```python
df.describe()
```

```python
df = df.dropna(subset=["Ride Distance"]).reset_index(drop=True)
```

```python
df = df.dropna(subset=["Driver Ratings"]).reset_index(drop=True)
```

```python
df = df.dropna(subset=["Customer Rating"]).reset_index(drop=True)
```

```python
df.isna().sum()
```

```python
lr = LabelEncoder()
codeVehicle = lr.fit_transform(df['Vehicle Type'])
```

```python
np.unique(codeVehicle)
```

```python
df['codeVehicle'] = codeVehicle
```

```python
import pandas as pd

# Convert Time column to datetime, parsing hours, minutes, seconds
df['Time_dt'] = pd.to_datetime(df['Time'], format='%H:%M:%S', errors='coerce')
```

```python
df['hour'] = df['Time_dt'].dt.hour
```

```python
def part_of_day(hour):
    if 5 <= hour < 12:
        return 0  # morning
    elif 12 <= hour < 17:
        return 1  # afternoon
    elif 17 <= hour < 21:
        return 2  # evening
    else:
        return 3  # night

df['part_of_day'] = df['hour'].apply(part_of_day)
```

```python
codePay = lr.fit_transform(df['Payment Method'])
```

```python
df['codePay'] = codePay
```

```python
df
```

```python
np.unique(codePay)
```

```python
df = df.reset_index(drop=True)
```

```python
dfx = df[['codeVehicle', 'Ride Distance','Avg VTAT','Avg CTAT']]
dfx.head()
```

```python
dfpca = dfx[['Ride Distance','Avg VTAT','Avg CTAT']]
```

```python
dfpca
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sl = StandardScaler()
dfpcas = sl.fit_transform(dfpca)
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA()
df_pca = pca.fit_transform(dfpcas)
```

```python
dfs = pd.DataFrame(df_pca)
```

```python
dfs.head()
```

```python
eigen_values = pca.explained_variance_
eigen_values
```

```python
variance = pca.explained_variance_ratio_
variance
```

```python
dfs.head()
```

```python
dfs['codeVehicle'] = dfx['codeVehicle']
dfs['time'] = df['part_of_day']
```

```python
dfs
```

```python
dfs_numpy = np.array(dfs)
```

```python
salary_numpy = np.array(df['Booking Value'])
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(dfs_numpy, salary_numpy, test_size = 0.2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr_model = LinearRegression()
model = lr_model.fit(xtrain, ytrain)
```

```python
model
```

```python
ypred = model.predict(xtest)
```

```python
from sklearn.metrics import mean_absolute_percentage_error, mean_absolute_error
```

```python
mape = mean_absolute_percentage_error(ypred, ytest)
print(f'{mape:.2%}')
```

```python
mae = mean_absolute_error(ypred, ytest)
print(f'{mae}')
```
